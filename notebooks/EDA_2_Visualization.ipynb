```python
import pandas as pd
```

```python
df = pd.read_csv("eda1_data_final.csv")
```

```python
df
```

```python
df = df.drop(columns="Unnamed: 0")
```

```python
df
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,10))

sns.countplot(x = "ACTIVITY", data = df)

plt.xlabel("Bioactivity_class", fontsize = 15, fontweight = "bold")
plt.ylabel("Frequency" , fontsize = 15 , fontweight = "bold")

plt.savefig("activity_freq_countplot.jpeg")
```

```python
plt.figure(figsize=(20,20))

sns.scatterplot(x='Mol_Wt', y= "LogP", data=df, hue='ACTIVITY', edgecolor='black', alpha=0.7)

plt.xlabel('Molecular_weight', fontsize=15, fontweight='bold')
plt.ylabel('Log_P', fontsize=15, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.savefig('MW_spread.jpeg')

#inactives and actives are pretty evenly distributed wrt molecular weight. This rules out mol wt as a prediction feature for the model. 
```

```python
plt.figure(figsize=(20,20))

sns.scatterplot(x='H_Donors', y= "H_acceptors", data=df, hue='ACTIVITY', edgecolor='black', alpha=0.7)

plt.xlabel('Hydrogen Donors', fontsize=15, fontweight='bold')
plt.ylabel('Hydrogen Acceptors', fontsize=15, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.savefig('hydrogen_spread.jpeg')
```
